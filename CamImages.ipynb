```
import matplotlib.pyplot as plt
import matplotlib.image as matIm
import os
import numpy as np
import sys
%matplotlib inline
```

```
blackList = ["RGB_16", "RGB_LSTA_16", "TwoStreamLSTA_16", "TwoStream_16", ".DS_Store"]
whiteList = ["RGB_7", "TwoStream_7", "self_sup", "self_sup_regr", "RGB_LSTA_7", "TwoStreamLSTA_7", "self_sup_flow"]
CAMs = "./CAMs 2"

stackSize = 5

action = "open_chocolate"
img_file = action + ".jpg"

x, y = 0,0

figure, axes = plt.subplots(7, 5, figsize=(3*5, 3*7))

for folder in whiteList:
  path = os.path.join(CAMs, folder, action)
  print(path)
  length = len(os.listdir(path))
  for i in np.linspace(1, length, stackSize):
    fl_name = path + "/1_" + "rgb" + str(int(np.floor(i))).zfill(4) + ".jpg"
    im = matIm.imread(fl_name)
    axes[x, y].imshow(im)
    axes[x,y].axis('off')
    #print(fl_name)
    y += 1
  x += 1
  y = 0

plt.subplots_adjust(wspace=0.02, hspace=0, bottom=0.05)
plt.savefig(img_file)
plt.show()
```
